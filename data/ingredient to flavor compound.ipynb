```python
chart = pd.read_csv("ingr_comp.tsv", sep='\t')
```

```python
chart['ingredient name'] = 'NaN'
```

```python
chart.head()
```

```python
for idx, row in chart.iterrows():
    ingr_id = int(chart['# ingredient id'].iloc[[idx]])
    try:
        chart.loc[idx,('ingredient name')] = ingredients.loc[ingr_id,('ingredient name')]
    except KeyError:
        pass
```
